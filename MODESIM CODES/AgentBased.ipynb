```python
people = virus.CreateWorld(300)
```

```python
# Infect the first 10 people
for i in range(10):
    people[i].infected = True
```

```python
NITER = 200
alive = []
infected = []
for i in range(NITER):
    virus.Iterate(people, 100, 100)
    ct = 0 # Counts people alive
    cu = 0 # Counts people infected
    for me in people:
        if me.alive:
            ct += 1
            if me.infected:
                cu += 1
    alive.append(ct)
    infected.append(cu)
```

```python
import matplotlib.pyplot as plt
plt.plot(alive, label = "Alive")
plt.plot(infected, label = "Infected")
```

```python
print("Number of Alive People at the end of 200 days: ", alive[-1])
print("Maximum Number of Infected People at the end of 200 days: ", max(infected))
```

```python
import numpy as np
```

```python
NPEOPLE = 300
NITERS = 300
M, B, I = virus.InitVirusMats(NPEOPLE)
B[:10, 1] = True # Infects the first 10 persons
inf = np.zeros(NITERS) # For Infected
aliv = np.zeros(NITERS) # For Alive
for i in range(NITERS):
    virus.IterateM(M, B, I)
    inf[i] = B[:,1].sum()
    aliv[i] = B[:,0].sum()
```

```python
import matplotlib.pyplot as plt
plt.plot(aliv, label = "Alive")
plt.plot(inf, label = "Infected")
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
```

```python
print("Number of Alive People at the end of 300 days: ", aliv[-1])
print("Maximum Number of Infected People during the simulation: ", max(inf))
```
